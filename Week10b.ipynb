```python
from collections import defaultdict
filename = "LabsCorePopulatedTable.csv"

grouped = defaultdict(list)
with open(filename) as file:
    header = next(file).strip().split("\t")
    for line in file:
        line = line.strip()

        line = dict(zip(header, line.strip().split("\t")))
        # group_by_field = f"{line.get("PatientID")}___{line.get("LabName")}"
        group_by_field = (line.get("PatientID"), line.get("LabName"))
        grouped[group_by_field].append(line)
```

```python
WITH AGE AS (
	SELECT 
	PATIENTID,
	FLOOR((JULIANDAY('NOW') - JULIANDAY(PATIENTDATEOFBIRTH))/365.25) AGE
	FROM 
	PATIENTCOREPOPULATEDTABLE
)
SELECT 
	PATIENTID,
	AGE, 
	CASE 
		WHEN AGE < 18 THEN 'YOUTH'
		WHEN AGE BETWEEN 18 AND 35 THEN 'YOUNG ADULT'
		WHEN AGE BETWEEN 36 AND 55 THEN 'ADULT'
		WHEN AGE >= 56 THEN 'SENIOR'
	END AGERANGE
FROM 
	AGE
ORDER BY AGE;
```
