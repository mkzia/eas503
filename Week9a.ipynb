```python
def func(line):
    name, *grades = line.split(',')
    grades = list(map(int, grades))
    average = sum(grades)/len(grades)
    return (name, average)

# aggregation and sorting
sorted(map(func, open(filename).readlines()), key=lambda ele: ele[1], reverse=True)
```
