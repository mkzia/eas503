* You input password into a text/input box
* This is sent securely encrypted to the server
* The server takes your password and runs a hashing algorithm and then checks the hash against what is stored in the database
* Main concept of a hash is that it is a one way street

```python
import bcrypt

password = b"firmflock95"
hashed_password = bcrypt.hashpw(password, bcrypt.gensalt())
print(hashed_password)
```

```python
import hashlib

data = b"firmflock95"
sha256_hash = hashlib.sha256(data).hexdigest()
print(sha256_hash)
```
