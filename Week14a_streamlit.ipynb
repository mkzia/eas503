## How to deploy streamlit app

* To deploy streamlit app, you need to commit your streamlit app to github.
* How do you commit your app to github?
    1. You need to install your SSH key in github
    2. Create a new repo on gitub.
    3. Associate your new repo with your local streamlit app folder
    4. Commit and push your streamlit app folder
    5. Now when you click on deploy, you will be guided to deploy your app.

### How do you install SSH key in Github?

#### Linux

* If you are using Linux container, execute command `ssh-keygen -t ed25519`. 
* This will create a .ssh folder inside your home folder `/root/.ssh`. 
* Open the file `id_ed25519.pub` and copy and paste its content into [github settings](https://github.com/settings/keys).

#### Windows

* For Windows, I recommend installing [gitbash](https://git-scm.com/install/windows).
* Launch gitbash and execute command `ssh-keygen -t ed25519`.
* This will create a .ssh folder inside your home folder, `/c/Users/mkhaw/.ssh`
* Open the file `id_ed25519.pub` and copy and paste its content into [github settings](https://github.com/settings/keys).
