```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/airline-safety/airline-safety.csv")
```

# Tasks
- Access with and without index
- List top 10 flights with the most km flown
- List bottom 10 flights
- Find 10 flights with the mosts number of incidents between 85-99
- For the above 10 flights, find their incidence rate between 00-14
- Do a bar plot for the above two tasks
- Reverse the above scenario
- Plot histogram of the above data
- Plot the mean and std -- error bars

```python
# List top 10 flights with the most km flown
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html
data.sort_values('avail_seat_km_per_week', ascending=False).head(10)
```

```python
data2 = data.sort_values('fatalities_85_99', ascending=False).head(10)
```

```python
data2[['airline', 'fatalities_85_99', 'fatalities_00_14']] 
```

```python
# List bottom 10 flights
data.sort_values('avail_seat_km_per_week', ascending=True).head(10)
```

```python
# Find 10 flights with the mosts number of incidents between 85-99
data2 = data.sort_values('incidents_85_99', ascending=False).head(10)
display(data2)
```

```python
# For the above 10 flights, find their incidence rate between 00-14
data2[['airline', 'incidents_85_99', 'incidents_00_14']] 
```

```python
# data2.set_index('airline', inplace=True)
data2 = data.sort_values('fatalities_85_99', ascending=False).head(10)
data2.set_index('airline', inplace=True)
display(data2)
data2 = data2[['fatalities_85_99', 'fatalities_00_14']] 
data2.plot.bar()
```
