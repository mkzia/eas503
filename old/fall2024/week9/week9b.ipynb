```python
filename = "PatientCorePopulatedTable.csv"
values = []
with open(filename) as file:
    next(file)
    for line in file:
        if not line.strip():
            continue
        _, _, _, _, mstatus, _, poverty = line.strip().split(',')
        values.append((mstatus, poverty))

sorted(values, key=lambda x: (x[0], float(x[1])))
```
