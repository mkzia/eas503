```python
"""
SELECT 
PatientID, 
Count(LabName) LabCount, 
Max(LabValue),  
Min(LabValue),  
SUM(LabValue), 
ROUND(AVG(LabValue),2) AVERAGE
FROM LabsCorePopulatedTable 
WHERE LabName = 'URINALYSIS: RED BLOOD CELLS'
GROUP BY PatientID
ORDER BY -AVERAGE
"""
from collections import defaultdict

filename = "LabsCorePopulatedTable.csv"
data = defaultdict(list)
with open(filename) as file:
    next(file)
    for line in file:
        tmp = line.strip().split(',')
        patient_id, lab_name, lab_value = tmp[0], tmp[2], tmp[3]
        if lab_name == 'URINALYSIS: RED BLOOD CELLS': # where
            data[patient_id].append(float(lab_value)) ###  # grouping

def avg(lst):
    return round(sum(lst)/len(lst), 2)

def format(patient_id, values):
    return (patient_id, len(values), max(values), min(values), sum(values), avg(values))


rows = []
for patient_id, values in data.items():
    rows.append(format(patient_id, values))

for row in sorted(rows, key=lambda row: row[-1], reverse=True):
    print(row)
```

```python
"""
SELECT 
PatientID, 
Count(LabName) LabCount, 
Max(LabValue),  
Min(LabValue),  
SUM(LabValue), 
ROUND(AVG(LabValue),2) AVERAGE
FROM LabsCorePopulatedTable 
WHERE LabName = 'URINALYSIS: RED BLOOD CELLS'
GROUP BY PatientID
ORDER BY -AVERAGE
"""
from collections import defaultdict

filename = "LabsCorePopulatedTable.csv"
data = defaultdict(dict)
with open(filename) as file:
    next(file)
    for line in file:
        tmp = line.strip().split(',')
        patient_id, lab_name, lab_value = tmp[0], tmp[2], tmp[3]
        if lab_name not in data[patient_id]:
            data[patient_id][lab_name] = []
        data[patient_id][lab_name].append(float(lab_value)) ###  # grouping

def avg(lst):
    return round(sum(lst)/len(lst), 2)

def format(patient_id, lab_name, values):
    return (patient_id, lab_name, len(values), max(values), min(values), sum(values), avg(values))


rows = []
for patient_id, labs in data.items():
    for lab_name in labs:
        values = data[patient_id][lab_name]
        rows.append(format(patient_id, lab_name, values))

for row in sorted(rows, key=lambda row: (row[0], -row[2])):
    print(row)

# 39	12	3.1	276.3	7.08
```
