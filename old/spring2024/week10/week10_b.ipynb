```python
values = []
with open('PatientCorePopulatedTable.csv') as file:
    header = next(file)
    for line in file:
        line = line.strip()
        if not line:
            continue

        line = line.split('\t')
        PatientMaritalStatus = line[4]
        PatientPopulationPercentageBelowPoverty = line[6]
        values.append((PatientMaritalStatus, PatientPopulationPercentageBelowPoverty))

sorted(values, key=lambda line: (line[0], -float(line[1])))
```
