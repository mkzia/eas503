```python
def sim_coin_toss(number_of_simulations):
    import random
    output = {}
#     output = {}.fromkeys(
#     ['HeadsCount', 'TailsCount'], 0)
    for i in range(number_of_simulations):
        toss = random.randint(0,1)
        
        if toss == 0:
            toss = 'Heads'
        else:
            toss = 'Tails'
            
        if toss not in output:
            output[toss] = 0
            
        output[toss] += 1
    
    return output
```

```python
no_tosses = 888888
data = sim_coin_toss(no_tosses)

group_data = list(data.values())
group_names = list(data.keys())

fig, ax = plt.subplots()
ax.bar(group_names, group_data)
ax.set(xlabel='Coins', ylabel='Frequency',
       title=f'Simulation with {no_tosses} Coin Tosses')
```
