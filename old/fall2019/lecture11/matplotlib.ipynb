```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(1, 10, 10)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Method 1
# A simple plot
plt.plot(x,y)
```

```python
# Shortcut for controlling the line style and color
plt.plot(x,y, 'r+')
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.plot.html
```

```python
# Adding labels and titles
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Subplots -- have multiple plots
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')

plt.subplot(1,2,2)
plt.plot(x,y, 'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Subplots -- have multiple plots
plt.subplot(2,2,1)
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('1')

plt.subplot(2,2,2)
plt.plot(x,y, 'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('2')

plt.subplot(2,2,3)
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('3')

plt.subplot(2,2,4)
plt.plot(x,y, 'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('4')
```

```python
# Method 2 -- Object Oriented way
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.5, 0.8])
# rect : sequence of float
# The dimensions [left, bottom, width, height] of the new axes. 
# All quantities are in fractions of figure width and height.
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1])
# rect : sequence of float
# The dimensions [left, bottom, width, height] of the new axes. 
# All quantities are in fractions of figure width and height.

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.12, 0.45, 0.45, 0.45])
# rect : sequence of float
# The dimensions [left, bottom, width, height] of the new axes. 
# All quantities are in fractions of figure width and height.

axes1.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Larger Plot')

axes2.plot(x,y, 'ro')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)


axes[0][0].plot(x,y)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)


for idx, ax in enumerate(axes.reshape(-1), 1): 
    ax.plot(x,y)
    ax.set_title(idx)
plt.tight_layout()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1])
# rect : sequence of float
# The dimensions [left, bottom, width, height] of the new axes. 
# All quantities are in fractions of figure width and height.

axes.plot(x,y, label='x,y')
axes.plot(x,y*3,label='x,y**3')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
axes.legend(loc=7, fontsize=24)
# axes.legend(loc=1)
# axes.legend(loc=[0.5,0.5])

# Location String 	Location Code
# 'best' 	0
# 'upper right' 	1
# 'upper left' 	2
# 'lower left' 	3
# 'lower right' 	4
# 'right' 	5
# 'center left' 	6
# 'center right' 	7
# 'lower center' 	8
# 'upper center' 	9
# 'center' 	10
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1])
# axes.plot(x,y, 'r')
# axes.plot(x,y, color='yellow')

# axes.plot(x,y, 'k')
axes.plot(x,y, color='black')

# # https://www.color-hex.com/
axes.plot(x,y, color='#a6b896')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1])
# axes.plot(x,y, 'r', linewidth=4)

axes.plot(x,y, 'r--o', lw=4, alpha=1.0)
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(x,y, 'r', lw=5, linestyle=':')
# Linestyle 	Description
# '-' or 'solid' 	solid line
# '--' or 'dashed' 	dashed line
# '-.' or 'dashdot' 	dash-dotted line
# ':' or 'dotted' 	dotted line
# 'None' or ' ' or '' 	draw nothing
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1])
# axes.plot(x,y, marker='o')
axes.plot(x,y, 'b', marker='o', markersize=10)
#https://matplotlib.org/3.1.1/api/markers_api.html
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1])
# axes.plot(x,y, 'b', marker='o', markersize=10, markerfacecolor='red')
axes.plot(x,y, 'b', marker='o', markersize=10, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='k')
#https://matplotlib.org/3.1.1/api/markers_api.html
```
