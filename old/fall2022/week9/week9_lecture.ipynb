```python
class ListV2:
   
    def __init__(self, values):
        self.values = values
        
    def __iter__(self):
        self.index = 0
        return self
    
    def __next__(self):
        if self.index < len(self.values):
            output = self.values[self.index]
            self.index += 1
            return output
        else:
            raise StopIteration
            
l1 = ListV2([1, 2, 3, 4])
```

```python
# iter(l1) ---> reset l1.__iter__
# next(l1) --> return values l1.__next__


i = iter(l1)
first_value = i.__next__()
print(first_value)
for val in i:
    print(val)
```

```python
filename = 'LabsCorePopulatedTable.csv'
from collections import defaultdict
data_dict = defaultdict(list)
header = None
lab_name = 'URINALYSIS: RED BLOOD CELLS'
with open(filename) as file:
    for line in file:
        if not line.strip():
            continue
            
#         if not header:
#             header = line.strip().split('\t')
#             continue
        values = line.strip().split('\t')
        
        if values[2] != lab_name: # WHERE LabName = 'URINALYSIS: RED BLOOD CELLS'
            continue
            
        patient_id = values[0]
        value = float(values[3])
        
        data_dict[patient_id].append(value)
        

def all_aggregations(patient_id, patient_lab_values):
    total = round(sum(patient_lab_values), 2)
    count = len(patient_lab_values)
    return (
        patient_id,
        count,
        round(max(patient_lab_values),2),
        round(min(patient_lab_values),2),
        total,
        round(total/count, 2),
    )
        
output = []
for patient_id in data_dict:
    output.append(all_aggregations(patient_id, data_dict[patient_id]))
```

```python
sorted(output, key=lambda ele: ele[-1], reverse=True)
```
