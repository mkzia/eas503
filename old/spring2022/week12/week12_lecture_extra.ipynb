```python
data = pd.read_csv('AdmissionsCorePopulatedTable.txt', delimiter='\t')
```

```python
data
```

```python
data['AdmissionStartDate'] = pd.to_datetime(data['AdmissionStartDate'])
data['AdmissionEndDate'] = pd.to_datetime(data['AdmissionEndDate'])
```

```python
data
```

```python
data['StayInDays'] = (data['AdmissionEndDate'] - data['AdmissionStartDate']).dt.days
```

```python
data.sort_values('AdmissionStartDate')
```

```python
data['AdmissionStartDate'].dt.year
```

```python
data
```

```python
data.groupby('PatientID').max().sort_values('StayInDays', ascending=False)
```
